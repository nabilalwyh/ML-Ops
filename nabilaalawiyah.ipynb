## Transform

Pada tahap ini, membuat modul exited.py yang berisi fungsi preprocessing untuk komponen Transform di TFX. Isinya mengatur cara mengolah data sebelum model dilatih: fitur numerik dikonversi ke float32, fitur biner dan label ke int64, serta nama kolom disesuaikan agar aman (menghapus spasi dan menambahkan akhiran _xf). Kolom Exited dijadikan label utama yang akan diprediksi model.

```python
TRANSFORM_MODULE_FILE = "exited.py"
```

```python
%%writefile {TRANSFORM_MODULE_FILE}
import tensorflow as tf

LABEL_KEY = "Exited"

# Helper untuk ubah nama kolom biar aman (hapus spasi)
def transformed_name(key):
    return key.replace(" ", "_") + "_xf"

NUMERIC_FEATURE_KEYS = [
    "CreditScore", "Age", "Tenure", "Balance", "NumOfProducts",
    "EstimatedSalary", "Satisfaction Score", "Point Earned"
]

BINARY_KEYS = ["HasCrCard", "IsActiveMember", "Complain"]

def preprocessing_fn(inputs):
    outputs = {}

    # Numeric features
    for key in NUMERIC_FEATURE_KEYS:
        safe_key = key.replace(" ", "_")
        outputs[transformed_name(key)] = tf.cast(inputs[key], tf.float32)

    # Binary features
    for key in BINARY_KEYS:
        outputs[transformed_name(key)] = tf.cast(inputs[key], tf.int64)

    # Label
    outputs[transformed_name(LABEL_KEY)] = tf.cast(inputs[LABEL_KEY], tf.int64)

    return outputs
```

Kode di atas menjalankan komponen Transform yang berfungsi untuk menerapkan proses preprocessing sesuai aturan yang sudah ditulis di file TRANSFORM_MODULE_FILE. Komponen ini mengambil data dari CsvExampleGen dan skema dari SchemaGen, lalu mengubah semua fitur mentah menjadi fitur yang sudah siap digunakan untuk pelatihan model (misalnya mengonversi tipe data atau menamai ulang kolom). Perintah interactive_context.run(transform) mengeksekusi proses tersebut dan menghasilkan transformed dataset yang akan digunakan oleh komponen Trainer pada tahap berikutnya.

```python
transform  = Transform(
    examples=example_gen.outputs['examples'],
    schema= schema_gen.outputs['schema'],
    module_file=os.path.abspath(TRANSFORM_MODULE_FILE)
)
interactive_context.run(transform)
```

## Trainer

### Membuat Tahapan Pengembangan Model

```python
# Fungsi transformed_name() untuk mengubah nama fitur yang telah melalui proses transform.

def transformed_name(key):
    """Renaming transformed features"""
    return key + "_xf"
```

```python
# Fungsi gzip_reader_fn() untuk memuat data dalam format TFRecord.
def gzip_reader_fn(filenames):
    """Loads compressed data"""
    return tf.data.TFRecordDataset(filenames, compression_type='GZIP')
```

```python
# Fungsi input_fn() untuk memuat transformed_feature yang dihasilkan oleh komponen Transform dan membaginya ke dalam beberapa batch.

def input_fn(file_pattern, 
             tf_transform_output,
             num_epochs,
             batch_size=64)->tf.data.Dataset:
    """Get post_tranform feature & create batches of data"""
    
    # Get post_transform feature spec
    transform_feature_spec = (
        tf_transform_output.transformed_feature_spec().copy())
    
    # create batches of data
    dataset = tf.data.experimental.make_batched_features_dataset(
        file_pattern=file_pattern,
        batch_size=batch_size,
        features=transform_feature_spec,
        reader=gzip_reader_fn,
        num_epochs=num_epochs,
        label_key = transformed_name(LABEL_KEY))
    return dataset
```

```python
# Fungsi model_builder() untuk membuat arsitektur model. 
from tensorflow.keras import layers

VOCAB_SIZE = 10000
SEQUENCE_LENGTH = 100
 
vectorize_layer = layers.TextVectorization(
    standardize="lower_and_strip_punctuation",
    max_tokens=VOCAB_SIZE,
    output_mode='int',
    output_sequence_length=SEQUENCE_LENGTH)
 
embedding_dim=16
def model_builder():
    """Build machine learning model"""
    inputs = tf.keras.Input(shape=(1,), name=transformed_name(FEATURE_KEY), dtype=tf.string)
    reshaped_narrative = tf.reshape(inputs, [-1])
    x = vectorize_layer(reshaped_narrative)
    x = layers.Embedding(VOCAB_SIZE, embedding_dim, name="embedding")(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dense(32, activation="relu")(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    
    model = tf.keras.Model(inputs=inputs, outputs = outputs)
    
    model.compile(
        loss = 'binary_crossentropy',
        optimizer=tf.keras.optimizers.Adam(0.01),
        metrics=[tf.keras.metrics.BinaryAccuracy()]
    
    )
    
    # print(model)
    model.summary()
    return model
```

```python
def _get_serve_tf_examples_fn(model, tf_transform_output):
    
    model.tft_layer = tf_transform_output.transform_features_layer()
    
    @tf.function
    def serve_tf_examples_fn(serialized_tf_examples):
        
        feature_spec = tf_transform_output.raw_feature_spec()
        
        feature_spec.pop(LABEL_KEY)
        
        parsed_features = tf.io.parse_example(serialized_tf_examples, feature_spec)
        
        transformed_features = model.tft_layer(parsed_features)
        
        # get predictions using the transformed features
        return model(transformed_features)
        
    return serve_tf_examples_fn
```

```python
# Fungsi run_fn() untuk menjalankan proses training model sesuai dengan parameter training yang diberikan.
from tfx.components.trainer.fn_args_utils import FnArgs

def run_fn(fn_args: FnArgs) -> None:
    
    log_dir = os.path.join(os.path.dirname(fn_args.serving_model_dir), 'logs')
    
    tensorboard_callback = tf.keras.callbacks.TensorBoard(
        log_dir = log_dir, update_freq='batch'
    )
    
    es = tf.keras.callbacks.EarlyStopping(monitor='val_binary_accuracy', mode='max', verbose=1, patience=10)
    mc = tf.keras.callbacks.ModelCheckpoint(fn_args.serving_model_dir, monitor='val_binary_accuracy', mode='max', verbose=1, save_best_only=True)
    
    
    # Load the transform output
    tf_transform_output = tft.TFTransformOutput(fn_args.transform_graph_path)
    
    # Create batches of data
    train_set = input_fn(fn_args.train_files, tf_transform_output, 10)
    val_set = input_fn(fn_args.eval_files, tf_transform_output, 10)
    vectorize_layer.adapt(
        [j[0].numpy()[0] for j in [
            i[0][transformed_name(FEATURE_KEY)]
                for i in list(train_set)]])
    
    # Build the model
    model = model_builder()
    
    
    # Train the model
    model.fit(x = train_set,
            validation_data = val_set,
            callbacks = [tensorboard_callback, es, mc],
            steps_per_epoch = 1000, 
            validation_steps= 1000,
            epochs=10)
    signatures = {
        'serving_default':
        _get_serve_tf_examples_fn(model, tf_transform_output).get_concrete_function(
                                    tf.TensorSpec(
                                    shape=[None],
                                    dtype=tf.string,
                                    name='examples'))
    }
    model.save(fn_args.serving_model_dir, save_format='tf', signatures=signatures)
```

Selanjutnya, membuat modul pelatihan model (Trainer module) untuk komponen Trainer di TFX. Fungsinya adalah mendefinisikan cara data hasil transformasi digunakan untuk membangun, melatih, dan menyimpan model machine learning.

```python
TRAINER_MODULE_FILE = "exited_trainer.py"
```

```python
%%writefile {TRAINER_MODULE_FILE}
import tensorflow as tf
import tensorflow_transform as tft 
from tensorflow.keras import layers
import os  
from tfx.components.trainer.fn_args_utils import FnArgs

LABEL_KEY = "Exited"

NUMERIC_FEATURE_KEYS = [
    "CreditScore", "Age", "Tenure", "Balance", "NumOfProducts",
    "EstimatedSalary", "Satisfaction Score", "Point Earned"
]

BINARY_KEYS = ["HasCrCard", "IsActiveMember", "Complain"]

# 💡 Gunakan fungsi yang sama seperti di transform.py
def transformed_name(key):
    return key.replace(" ", "_") + "_xf"

def gzip_reader_fn(filenames):
    return tf.data.TFRecordDataset(filenames, compression_type='GZIP')

def input_fn(file_pattern, tf_transform_output, num_epochs, batch_size=64) -> tf.data.Dataset:
    transform_feature_spec = tf_transform_output.transformed_feature_spec().copy()
    dataset = tf.data.experimental.make_batched_features_dataset(
        file_pattern=file_pattern,
        batch_size=batch_size,
        features=transform_feature_spec,
        reader=gzip_reader_fn,
        num_epochs=num_epochs,
        label_key=transformed_name(LABEL_KEY)
    )
    return dataset.repeat()  # 🩹 tambahkan repeat agar tidak "ran out of data"

def model_builder(tf_transform_output):
    """Build ML model with numeric, binary, and categorical features"""

    # Input layers untuk semua fitur
    inputs = {}
    for key in NUMERIC_FEATURE_KEYS + BINARY_KEYS:
        inputs[transformed_name(key)] = tf.keras.Input(shape=(1,), name=transformed_name(key))

    # Proses numeric features
    numeric_processed = [tf.cast(inputs[transformed_name(key)], tf.float32) for key in NUMERIC_FEATURE_KEYS]

    # Proses binary features
    binary_processed = [tf.cast(inputs[transformed_name(key)], tf.float32) for key in BINARY_KEYS]

    # Gabungkan semua feature
    all_features = tf.keras.layers.concatenate(numeric_processed + binary_processed)

    # Dense layers
    x = layers.Dense(64, activation='relu')(all_features)
    x = layers.Dense(32, activation='relu')(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    model.compile(
        loss='binary_crossentropy',
        optimizer=tf.keras.optimizers.Adam(0.01),
        metrics=[tf.keras.metrics.BinaryAccuracy()]
    )
    model.summary()
    return model

def _get_serve_tf_examples_fn(model, tf_transform_output):
    model.tft_layer = tf_transform_output.transform_features_layer()
    
    @tf.function
    def serve_tf_examples_fn(serialized_tf_examples):
        feature_spec = tf_transform_output.raw_feature_spec()
        feature_spec.pop(LABEL_KEY)
        parsed_features = tf.io.parse_example(serialized_tf_examples, feature_spec)
        transformed_features = model.tft_layer(parsed_features)
        return model(transformed_features)
    
    return serve_tf_examples_fn

def run_fn(fn_args: FnArgs) -> None:
    log_dir = os.path.join(os.path.dirname(fn_args.serving_model_dir), 'logs')
    tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, update_freq='batch')
    es = tf.keras.callbacks.EarlyStopping(monitor='val_binary_accuracy', mode='max', verbose=1, patience=10)
    mc = tf.keras.callbacks.ModelCheckpoint(fn_args.serving_model_dir, monitor='val_binary_accuracy', mode='max', verbose=1, save_best_only=True)
    
    # Load Transform output
    tf_transform_output = tft.TFTransformOutput(fn_args.transform_graph_path)
    
    # Create train & eval datasets
    train_set = input_fn(fn_args.train_files, tf_transform_output, 10)
    val_set = input_fn(fn_args.eval_files, tf_transform_output, 10)
    
    # Build the model
    model = model_builder(tf_transform_output)
    
    # Train
    model.fit(
        x=train_set,
        validation_data=val_set,
        callbacks=[tensorboard_callback, es, mc],
        steps_per_epoch=1000,
        validation_steps=1000,
        epochs=10
    )
    
    # Save model with serving signature
    signatures = {
        'serving_default': _get_serve_tf_examples_fn(model, tf_transform_output).get_concrete_function(
            tf.TensorSpec(shape=[None], dtype=tf.string, name='examples'))
    }
    
    model.save(fn_args.serving_model_dir, save_format='tf', signatures=signatures)
```

Data yang digunakan berasal dari output komponen Transform, yaitu transformed_examples (data yang sudah diproses) dan transform_graph (graf pemetaan transformasi). Selain itu, schema_gen.outputs['schema'] digunakan untuk memastikan struktur data sesuai. Parameter train_args dan eval_args menentukan pembagian data untuk pelatihan (train) dan evaluasi (eval). Perintah interactive_context.run(trainer) kemudian mengeksekusi seluruh proses training sesuai konfigurasi tersebut hingga menghasilkan model terlatih yang siap dievaluasi dan disimpan.

## Resolver

Komponen Resolver digunakan agar pipeline dapat mengidentifikasi model terbaru yang sudah lolos evaluasi (blessed) untuk dibandingkan dengan model baru yang akan dilatih. Channel(type=Model) menunjuk pada artefak model sebelumnya, sedangkan Channel(type=ModelBlessing) menunjuk pada hasil evaluasinya. Dengan strategi LatestBlessedModelStrategy, pipeline secara otomatis memilih model terbaru yang telah “diberkati” sebagai acuan (baseline model) dalam proses evaluasi berikutnya. Perintah interactive_context.run(model_resolver) kemudian mengeksekusi proses pencarian model tersebut agar dapat digunakan pada tahap Evaluator.

```python
from tfx.dsl.components.common.resolver import Resolver 
from tfx.dsl.input_resolution.strategies.latest_blessed_model_strategy import LatestBlessedModelStrategy 
from tfx.types import Channel 
from tfx.types.standard_artifacts import Model, ModelBlessing 
 
model_resolver = Resolver(
    strategy_class= LatestBlessedModelStrategy,
    model = Channel(type=Model),
    model_blessing = Channel(type=ModelBlessing)
).with_id('Latest_blessed_model_resolver')
 
interactive_context.run(model_resolver)
```

## Evaluator

Konfigurasi ini menentukan cara model dievaluasi dan metrik apa saja yang dihitung. Bagian model_specs menunjukkan bahwa label target yang digunakan untuk evaluasi adalah kolom Exited. Parameter slicing_specs=[tfma.SlicingSpec()] berarti evaluasi dilakukan pada seluruh data tanpa pemisahan berdasarkan fitur tertentu (bisa ditambahkan nanti jika ingin melihat performa per segmen).

Pada bagian metrics_specs, terdapat daftar metrik evaluasi yang digunakan, yaitu:

* ExampleCount, jumlah total sampel yang dievaluasi,
* AUC, mengukur kemampuan model membedakan kelas positif dan negatif,
* FalsePositives, TruePositives, FalseNegatives, TrueNegatives untuk melihat hasil klasifikasi model secara detail,
* BinaryAccuracy, metrik utama dengan ambang batas (threshold) minimal 0.5 dan perubahan kinerja model sebelumnya harus meningkat setidaknya 0.0001 agar dianggap lebih baik.

Konfigurasi ini nantinya akan digunakan pada komponen Evaluator untuk menilai kualitas model yang telah dilatih.

```python
import tensorflow_model_analysis as tfma 
 
eval_config = tfma.EvalConfig(
    model_specs=[tfma.ModelSpec(label_key='Exited')],
    slicing_specs=[tfma.SlicingSpec()],
    metrics_specs=[
        tfma.MetricsSpec(metrics=[
            
            tfma.MetricConfig(class_name='ExampleCount'),
            tfma.MetricConfig(class_name='AUC'),
            tfma.MetricConfig(class_name='FalsePositives'),
            tfma.MetricConfig(class_name='TruePositives'),
            tfma.MetricConfig(class_name='FalseNegatives'),
            tfma.MetricConfig(class_name='TrueNegatives'),
            tfma.MetricConfig(class_name='BinaryAccuracy',
                threshold=tfma.MetricThreshold(
                    value_threshold=tfma.GenericValueThreshold(
                        lower_bound={'value':0.5}),
                    change_threshold=tfma.GenericChangeThreshold(
                        direction=tfma.MetricDirection.HIGHER_IS_BETTER,
                        absolute={'value':0.0001})
                    )
            )
        ])
    ]
 
)
```

```python
from tfx.components import Evaluator
evaluator = Evaluator(
    examples=example_gen.outputs['examples'],
    model=trainer.outputs['model'],
    baseline_model=model_resolver.outputs['model'],
    eval_config=eval_config)
 
interactive_context.run(evaluator)
```

## Pusher

Komponen Pusher mengambil model hasil pelatihan dari trainer.outputs['model'] dan hasil evaluasinya dari evaluator.outputs['blessing']. Jika model telah “diberkati” (artinya lolos evaluasi dan memenuhi ambang batas performa), maka Pusher akan menyalin model tersebut ke direktori yang ditentukan dalam push_destination. Pada kode ini, model disimpan ke folder /tmp/tfx_pipeline/serving_model_dir, sehingga siap digunakan untuk tahap selanjutnya seperti serving atau deployment di lingkungan produksi. Perintah interactive_context.run(pusher) mengeksekusi proses push tersebut.

```python
from tfx.components import Pusher
from tfx.proto import pusher_pb2

pusher = Pusher(
    model=trainer.outputs['model'],
    model_blessing=evaluator.outputs['blessing'],
    push_destination=pusher_pb2.PushDestination(
        filesystem=pusher_pb2.PushDestination.Filesystem(
            base_directory='/tmp/tfx_pipeline/serving_model_dir'
        )
    )
)

interactive_context.run(pusher)
```
